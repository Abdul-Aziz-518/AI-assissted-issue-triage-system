```python
import pandas as pd
from datetime import datetime
df=pd.read_csv("C:/Users/abdul/OneDrive/Desktop/AI assissted issue triage system/data/issues.csv", index_col="TicketID")
def _issue():
    ticket_id = f"TCKT-{len(df) + 1:04d}"
    name = input("Enter your name: ")
    roll = input("Enter your Roll Number: ")
    location = input("Enter the location of the issue: ").lower()
    issue_desc = input("Enter your issue in detail: ").lower()
    priority = pri(location, issue_desc)
    df.loc[ticket_id] = {
        "Name": name,
        "RollNumber": roll,
        "Location": location,
        "IssueDescription": issue_desc,
        "Date": datetime.now().date(),
        "Time": datetime.now().time(),
        "Status": "Open",
        "Priority": priority
    }

    df.to_csv(
        "C:/Users/abdul/OneDrive/Desktop/AI assissted issue triage system/data/issues.csv"
    )

def pri(l,i):
    l1=["staircase","hostle","lift","canteen","washroom","lab"]
    l2=["classroom","ground","hall"]
    i1=["malfunctioning","water leakage","unstable","loose","overheating","lift","unhygenic"]
    i2=["not working","not able", "not functioning","not good","bad"]
    if any(loc in l for loc in l1) and any(des in i for des in i1):
        return 5
    elif any(des in i for des in i1):
        return 4
    elif any(loc in l for loc in l1):
        return 3
    elif any(loc in l for loc in l2) or any(des in i for des in i2):
        return 2
    else:
        return 1
def update():
    t=input("Enter your complete ticketID: ")
    if t in df.index:
        print("Current Stautus:", df.loc[t]["Status"])
        c=input("Do you want to update the current status. Enter only=(Y/N)")
        if c=="Y":
            if df.loc[t]["Status"]=="Open":
                df.loc[t,"Status"]="In Progress"
            elif df.loc[t]["Status"]=="In Progress":
                df.loc[t,"Status"]="Closed"
            else:
                print("Status of your ticket:",df.loc[t,"Status"])
                print("Thank You")
            print("Status Updated. Current Status:", df.loc[t]["Status"])
        elif c=="N":
            print("Thank You")
        else:
            print("Invalid entry")
    else:
        print("Invalid entry, No such Id exists")
    df.to_csv("C:/Users/abdul/OneDrive/Desktop/AI assissted issue triage system/data/issues.csv")
```
